# Python Fundamentals
This notebook demonstrates how to perform a variety of fundamental tasks in the Python programming language.

### How to Display Text in Python

```python
#display a line of text using the  print()  function
print('Hello, my name is omar!')
```

### How to Include a Linefeed when Displaying Text in Python

```python
#print two lines of text by using the new line symbol
print('hello , my name is omar \n my age is 21')
```

### How to Declare Variables in Python

```python
# Defining a string variable to store a person's name
name = "omar"  # 'name' holds a sequence of characters (string)

# Defining an integer variable to store the person's age
age = 30  # 'age' holds an integer value representing years

# Defining a float variable to store the person's height in meters
height = 1.75  # 'height' holds a floating-point number (decimal value)

# Defining a boolean variable to indicate if the person is a student
is_student = False  # 'is_student' holds a boolean value: True or False

# Printing the values and their types
print("Name:", name)               # Output: Name: omar
print("Type of name:", type(name)) # Output: Type of name: <class 'str'>

print("Age:", age)                 # Output: Age: 30
print("Type of age:", type(age))   # Output: Type of age: <class 'int'>

print("Height:", height)           # Output: Height: 1.75
print("Type of height:", type(height)) # Output: Type of height: <class 'float'>

print("Is Student:", is_student)       # Output: Is Student: False
print("Type of is_student:", type(is_student)) # Output: Type of is_student: <class 'bool'>
```

### Arithmetic Operators in Python

```python
#set the value of a variable named 'a' to 5
a = 5

#addition. Result: 7
b = a + 2

#subtraction. Result: 4
c = b - 3

#multiplication. Result: 8
d = c * 2

#division. Result: 2.0
e = d / 4

#exponentiation. Result: 4.0
f = e ** 2

#modulos (returns the remainder). Result: 0.0
g = f % 2

#print results
print('The values of a, b, c, d, e, f, and g are: {0}, {1}, {2}, {3}, {4}, {5}, and {6}.'.format(a, b, c, d, e, f, g))
```

#Indexing:
Accesses individual characters using an index (starts from 0).

```python
word = "Python"
print(word[0])   # Output: P (first character)
print(word[-1])  # Output: n (last character)
```

#Slicing:

```python

word = "Python"
print(word[0:3])   # Output: Pyt (characters from index 0 to 2)

```
